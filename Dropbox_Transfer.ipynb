```python
import os
import dropbox

# Dropbox access token 
ACCESS_TOKEN = '####'

# Function to download files from Dropbox folder
def download_folder(dbx, folder_path, local_path):
    # List files in the folder
    try:
        folder_metadata = dbx.files_list_folder(folder_path)
    except dropbox.exceptions.ApiError as err:
        print(f'Error listing folder: {err}')
        return

    # Download each file in the folder
    for entry in folder_metadata.entries:
        if isinstance(entry, dropbox.files.FileMetadata):
            file_path = entry.path_lower
            local_file_path = os.path.join(local_path, os.path.basename(file_path))

            print(f'Downloading {file_path} to {local_file_path}')

            try:
                with open(local_file_path, 'wb') as f:
                    metadata, res = dbx.files_download(file_path)
                    f.write(res.content)
            except dropbox.exceptions.ApiError as err:
                print(f'Error downloading file {file_path}: {err}')

def main():
    # Initialize Dropbox client
    dbx = dropbox.Dropbox(ACCESS_TOKEN)

    # Specify Dropbox folder path to download
    dropbox_lateral_folder_path = '/Apps/PNGS_Spine/Lateral' 
    dropbox_oblique_folder_path = '/Apps/PNGS_Spine/Oblique' 
    dropbox_AP_folder_path = '/Apps/PNGS_Spine/Anterior_Posterior'
    
    # Specify local directory to save files
    local_lateral_download_path = '/home/jpatel6/Orientation_Data/Lateral'  
    local_oblique_download_path = '/home/jpatel6/Orientation_Data/Oblique'  
    local_AP_folder_path = '/home/jpatel6/Orientation_Data/AP'  

    # Download the Dropbox folder
    download_folder(dbx, dropbox_lateral_folder_path, local_lateral_download_path)
    download_folder(dbx,dropbox_oblique_folder_path, local_oblique_download_path)
    download_folder(dbx, dropbox_AP_folder_path, local_AP_folder_path)

    print('Download complete.')

if __name__ == '__main__':
    main()
```
